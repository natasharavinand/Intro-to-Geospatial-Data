# 2. Geometry Functions

```python
from shapely.geometry import Point, LineString, Polygon, MultiPoint
```

```python
# create Point Geometries
point1 = Point(2.04, 45.31)
point2 = Point(55.70, 11.97)
point3 = Point(40.59, -74.12)
point4 = Point(-15.81, -47.87)

# create LineString
line1 = LineString([point1, point2, point3])
line2 = LineString([point2, point3, point4])

# create a polygon
poly = Polygon([(-35, 0), (-35, 60), (60, 60), (60, 0), (0,0)])
```

## 1.1 Distance

```python
point1.distance(point2) # in degrees
```

```python
point4.distance(point1)
```

```python
point4.distance(line1) # can also find distance from a point to a line
```

## 1.2 Buffer

Buffering usually creates two areas: one area that is within a specified distance to selected real world features and the other area that is beyond. The area that is within the specified distance is called the buffer zone.

```python
point1.buffer(50) # by 50 decimal degrees
```

```python
MultiPoint([point1, point2, point3, point4]).buffer(10)
```

```python
MultiPoint([point1, point2, point3, point4]).buffer(50)
```

```python
line1.buffer(10)
```

## 1.3 Centroid

```python
# the centroid of a point is the same as a point

point1.centroid
```

```python
point1
```

```python
# the centroid of the line is the centermost point of the line

line1.centroid
```

```python
line1
```

```python
# we most often use centroids with polygons

poly.centroid
```

```python
poly
```

## 1.4 Binary Predicates

### 1.4.1 Contains

```python
poly.contains(point1)
```

```python
poly.contains(point4)
```

### 1.4.2 Within

```python
point1.within(poly)
```

```python
point4.within(poly)
```

```python
poly.within(point1)
```

### 1.4.3 Crosses

```python
line1.crosses(poly)
```

```python
line2.crosses(poly)
```

```python
point1.crosses(poly)
```

## 1.5 Set-Theoretic Methods

### 1.5.1 Difference

```python
point1_buffer = point1.buffer(60) # these are polygons
point2_buffer = point2.buffer(60)
```

```python
point1_buffer
```

```python
point2_buffer
```

```python
MultiPoint([point1, point2])
```

```python
MultiPoint([point1, point2]).buffer(60)
```

```python
point1_buffer.difference(point2_buffer) # point1_buffer - point2_buffer
```

```python
point2_buffer.difference(point1_buffer) # point2_buffer - point1_buffer
```

### 1.5.2 Intersection

```python
point1_buffer.intersection(point2_buffer)
```

### 1.5.3 Union

```python
point1_buffer.union(point2_buffer)
```

### 1.5.4 Symmetric Difference

```python
point1_buffer.symmetric_difference(point2_buffer) # only areas that the two polygons don't share
```
